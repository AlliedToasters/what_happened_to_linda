<H1>What Happened to Linda?</H1>
<H3>What's the story behind America's trendiest name?</H3>

This investigation begins with the public dataset of US Social Security applications, <A href="https://cloud.google.com/bigquery/public-data/usa-names">searchable via Google BigQuery</A>. My first query is all baby names that occured more than 500 times in each year in the dataset, 1910 to 2016.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('popular_names.csv') #The most popular baby names for each given year, 1910-2016
df['given_name'] = df['name'] #to avoid ambiguity
```

I want a list of all the names that were most popular in any year in this set:

```python
years = range(1910, 2017)
popular_names = []
for year in years:
    name = df.loc[df[df.year == year].number.idxmax()].given_name #get the most popular name for every given year
    popular_names.append(name)
popular_names = set(popular_names) #The set of all names that were most popular in any year, 1910-2016
print('the number of names that were most popular in any given year, 1910-2016, is: {}'.format(len(popular_names)))
```

<H1>The Problem</H1>
<H2>What Happened with the Name, "Linda?"</H2>

What drove America to name the highest number of babies the same name in a given year "Linda" in 1947? A quick Google search of "linda 1947" returns <A href='https://en.wikipedia.org/wiki/Linda_(1946_song)'>this wikipedia article</A> about a 1946 Buddy Clark tune, "Linda." Along with it come a lot of <A href='http://myhistoryfix.com/entertainment/linda-name-generation/'>speculations</A> about how this particular song influenced naming behavior in this period. 

Speculators assert that, driven by the popularity of this one song, American parents went into a "Linda" naming frenzy. It's a compelling story, but I was unable to find a supporting body of data that conclusively links this naming trend to this song.

<H1>Proposed Solution: The Linda Survey</H1>

A survey of women named Linda who were born in the US and hold Social Security numbers in the period of Linda's highest popularity as a name, from 1945 to 1950, will provide more conclusive data about the source of this naming anomaly. 

<H1>Method</H1>

<H2>Threshold for Success</H2>

The popularity of any name is a tricky subject. This particular question is complicated by another song entitled "Linda," released on the soundtrack of the 1945 film <I><A href='https://en.wikipedia.org/wiki/The_Story_of_G.I._Joe'>The Story of G.I. Joe</A></I>, although I can't find any references to a character named Linda in the film (which could have also influenced naming behavior.) 

Additionally, "Linda" appears to have been gaining popularity independently in preceding years.

```python
total_babies = pd.read_csv('total_babies.csv') #requeried db to get total babies born each year
total_babies['number_Linda'] = list(df[df.given_name == 'Linda'].sort_values('year').number)
total_babies['proportion_Linda'] = total_babies.number_Linda/total_babies.total_babies
```

```python
fig = plt.figure(figsize=(12,6))
ax = sns.tsplot(time=total_babies.year, data=total_babies.proportion_Linda);
ax.set_title('Proportion of US Babies Named "Linda" by Year');
plt.axvline(x=1945, label='1945, Release of G.I. Joe Film', linestyle='dotted', color='g');
plt.axvline(x=1947, label='1947, Buddy Clark song hits Billboard charts', linestyle='dotted', color='r');
plt.legend();
```

It appears that the 1945 <I>The Story of G.I. Joe</I> release could have influenced naming behavior, but it's not clear; the name was steadily gaining popularity already in 1945.

To account for the general popularity of the name, we'll look at the proportion of babies named Linda in 1944 and at its peak in 1948 (while more babies were named Linda in 1947, the proportion was highest in 1948.) The ratio of these two proportions could be a good target ratio for an affirmative result.

```python
song_name_ratio = float(total_babies[total_babies.year == 1948].proportion_Linda)/float(total_babies[total_babies.year == 1944].proportion_Linda)
print(str(song_name_ratio)[:4])
print(str(1/song_name_ratio)[:4])
```

By this measure, we might expect about half of the babies named Linda in 1948 to have been named after either song entitled "Linda."

<H2>Selection and Distribution</H2>


To isolate some variables in naming behavior, we'll distribute the survey to a portion of all Americans born in the period of 1945-1950. 

```python
population_size = total_babies[total_babies.year >= 1945][total_babies.year <= 1950].total_babies.sum()
linda_population = total_babies[total_babies.year >= 1945][total_babies.year <= 1950].number_Linda.sum()
print('the population size is: {}'.format(population_size))
print('the Linda population size is: {}'.format(linda_population))
```

There are 18,885,899 people in this population.
Using <A href='https://www.surveymonkey.com/mp/sample-size-calculator/'>this simple sample size calculator</A> from SurveyMonkey, a good sample of this population less than 10,000 people.

Although random selection is ideal, selecting and finding people is a difficult task. Instead, we'll reach out to AARP, as most of the surving members of our population of interest are probably retired. We'll sponsor the outreach and pay for a free year of membership for every respondent. With a discounted annual membership fee of \$12.60 per year, this will cost \$126,000. 

AARP will reach out to its members electronically and distribute the electronic survey to the first 10,000 respondents (we'll solicit 10,000 responses so we can reject invalid responses and still reach our target sample size.) This limits our selection to surviving AARP members who are interested in our incentive and respond fastest, but we'll assume that these variables have no correlation to naming behavior.

<H2>Request for Participation Verbage</H2>

To prevent self-selection based on naming behavior, we will not mention the subject of our study in the outreach email distributed by AARP. We will use their database to distribute these emails to the members in our age group that are US-born.

<H2>The Survey</H2>

The survey is simple but addresses our question, and lets us confirm the validity of each subject. We'll work with AARP to make sure all the respondents qualify as being born in the right period and being registered with the Social Security Service, but include these qualifiers in the survey for certainty.

<H3>Name Survey</H3>

1. What was your given first name at birth? (Input field: text)
2. Were you born in the US and registered with in the US Social Security system? (Check box: yes, no)
3. What is your Date of Birth? (Input field: dd/mm/yyyy)
4. Do you know why your given first name was chosen? (Check box: yes, no)
5. If so, what's the reason? (Input field: text)

<H2>Analysis</H2>

```python
print('the proportion of our population named Linda is: {}'.format(str(linda_population/population_size)[:5]))
```

In our method we assume name and naming behavior has no influence on sampling. If this is true, we should expect about 2.4%, or 240 respondents, to be named Linda. If this is not the case, our selection may be skewed and some of our assumptions may be incorrect.

<H4>Control Group: Respondents Not Named Linda</H4>
<H4>Treatment Group: Respondents Named Linda</H4>
<H4>Hypothesis:</H4> A higher proportion of the treatment group will have included "song" in their reason for naming than the control group.
<H4>Null Hypothesis:</H4>Being named Linda has no correlation with including "song" in the reason for naming response.

We'll use a simple regular expression to find all respondents who included the word "song" in their response to question five. We'll use the control group to identify the "standard" proportion of people who reference "song" in their answer to question five, and expect the treatment group proportion to be .5 greater than the control.

For rigor, we'll apply the z-test, appropriate for dealing with population proportions such as these. This may or may not let us rule out the null hypothesis and will tell us if our result is statistically significant.
