```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import colorsys
%matplotlib inline
```

```python
df = pd.read_csv('michaelvsmilo.csv')
```

```python
plt.figure(figsize=(8,6));
plt.plot(df[df.name=='Milo'].year, df[df.name=='Milo'].number, label='Milo');
plt.plot(df[df.name=='Michael'].year, df[df.name=='Michael'].number, label='Michael');
plt.title('US Babies Named Michael vs. Milo, 1910-2016');
plt.legend();

```

```python
df = pd.read_csv('popular_names.csv')
df['given_name'] = df['name']
```

```python
years = range(1910, 2017)
popular_names = []
for year in years:
    #popular_names.update(df.loc[df[df.year == year].number.idxmax()].given_name)
    name = df.loc[df[df.year == year].number.idxmax()].given_name
    popular_names.append(name)
popular_names = set(popular_names)
len(popular_names)
```
